```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")
model.train(
    data="/content/drive/MyDrive/Syook/datasets/data_split/data.yml",
    epochs=100,
    batch=16,
    imgsz=640,
    device="cuda"
)
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
class_names = [
    "person", "hard-hat", "gloves", "mask", "glasses",
    "boots", "vest", "ppe-suit", "ear-protector", "safety-harness"
]

# Define unique colors for each class
class_colors = {
    "person": (255, 0, 0),  # Blue
    "hard-hat": (0, 255, 0),  # Green
    "gloves": (0, 0, 255),  # Red
    "mask": (255, 255, 0),  # Cyan
    "glasses": (255, 0, 255),  # Magenta
    "boots": (0, 255, 255),  # Yellow
    "vest": (128, 0, 128),  # Purple
    "ppe-suit": (128, 128, 0),  # Olive
    "ear-protector": (0, 128, 128),  # Teal
    "safety-harness": (255, 165, 0)  # Orange
}

# Load and process the image
image_path = "/content/drive/MyDrive/Syook/datasets/images/-2091-_png_jpg.rf.24a38225fa17a89f450e6fcf90584bb5.jpg"
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Run inference
results = model(image_rgb)

# Draw bounding boxes
for result in results:
    for box, conf, cls in zip(result.boxes.xyxy, result.boxes.conf, result.boxes.cls):
        x1, y1, x2, y2 = map(int, box)
        class_name = class_names[int(cls)]
        color = class_colors.get(class_name, (255, 255, 255))  # Default white if not found

        # Draw bounding box
        cv2.rectangle(image_rgb, (x1, y1), (x2, y2), color, 2)

        # Display label and confidence
        label = f"{class_name} {conf:.2f}"
        text_size, _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 2)
        text_w, text_h = text_size
        cv2.rectangle(image_rgb, (x1, y1 - text_h - 5), (x1 + text_w, y1), color, -1)  # Text background
        cv2.putText(image_rgb, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 0), 2)

# Display the image
plt.figure(figsize=(10, 8))
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

```python
import os

label_path = "/content/drive/MyDrive/Syook/datasets/labels"
class_counts = {}

for file in os.listdir(label_path):
    with open(os.path.join(label_path, file), 'r') as f:
        for line in f.readlines():
            class_id = int(line.split()[0])
            class_counts[class_id] = class_counts.get(class_id, 0) + 1

print("Class Distribution:", class_counts)
```

```python
import matplotlib.pyplot as plt

# Class Distribution and Names
class_distribution = {0: 1284, 1: 972, 2: 447, 5: 895, 6: 194, 7: 240, 3: 8}
class_names = {
    0: 'person',
    1: 'helmet',
    2: 'vest',
    3: 'mask',
    5: 'gloves',
    6: 'boots',
    7: 'goggles'
}

# Sort by class name (optional for consistent order)
sorted_items = sorted(class_distribution.items(), key=lambda x: class_names.get(x[0], f'Class {x[0]}'))
classes = [class_names[k] for k, _ in sorted_items]
counts = [v for _, v in sorted_items]

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(classes, counts, color='skyblue')

# Add labels on top of bars
for bar, count in zip(bars, counts):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, str(count), ha='center', va='bottom')

# Customize plot
plt.xlabel("Class")
plt.ylabel("Number of Samples")
plt.title("Class Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
